# POPULARITY BASED RECOMENDER SYSTEM


```python
ratings_with_hotel_names = ratings.merge(hotels,on ='Hotel_ID')
```

```python
ratings_with_hotel_names
```

```python
#Count of average score for each hotels
num_of_scorings = ratings_with_hotel_names.groupby('Hotel_Name').count()['Additional_Number_of_Scoring'].reset_index()
num_of_scorings.rename(columns={'Additional_Number_of_Scoring':'count_Additional_Number_of_Scoring'},inplace = True)
num_of_scorings
```

```python
avg_num_of_scorings = ratings_with_hotel_names.groupby('Hotel_Name').mean()['Additional_Number_of_Scoring'].reset_index()
avg_num_of_scorings.rename(columns={'Additional_Number_of_Scoring':'Avg_Additional_Number_of_Scoring'},inplace = True)
avg_num_of_scorings
```

```python
popular_hotels = num_of_scorings.merge(avg_num_of_scorings, on = 'Hotel_Name')
popular_hotels
```

```python
popular_hotels = popular_hotels[popular_hotels['Avg_Additional_Number_of_Scoring']>=500].sort_values('Avg_Additional_Number_of_Scoring', ascending =False).head(50)
popular_hotels
```

```python
popular_hotels.merge(hotels, on='Hotel_Name').drop_duplicates('Hotel_Name')[['Hotel_ID', 'Avg_Additional_Number_of_Scoring', 'Hotel_Name', 'Hotel_Address','Tags' ]]
```

# Colleborative Filtering Based Recomendation system

```python
ratings_with_reviewers = ratings.merge(hotels,on ='Hotel_ID').merge(reviewers, on='Hotel_ID')#.count()['Total_Number_of_Reviews_Reviewer_Has_Given']#.reset_index()
ratings_with_reviewers
```

```python
positive_rating_of_reviewers = ratings_with_reviewers.groupby(['Hotel_Name', 'Average_Score', 'Reviewer_Score','Reviewer_Nationality']).count()['Total_Number_of_Reviews_Reviewer_Has_Given'].reset_index()
positive_rating_of_reviewers.rename(columns={'Total_Number_of_Reviews_Reviewer_Has_Given':'count_Total_Number_of_Reviews_Reviewer_Has_Given'},inplace = True)
positive_rating_of_reviewers = positive_rating_of_reviewers[(positive_rating_of_reviewers['count_Total_Number_of_Reviews_Reviewer_Has_Given'] > 50) & (positive_rating_of_reviewers['Reviewer_Score'] > 6.0) & (positive_rating_of_reviewers['Average_Score'] > 6.0)].sort_values('Reviewer_Score', ascending =False)
positive_rating_of_reviewers
```

```python
negative_rating_of_reviewers = ratings_with_reviewers.groupby(['Hotel_Name', 'Average_Score', 'Reviewer_Score','Reviewer_Nationality']).count()['Total_Number_of_Reviews_Reviewer_Has_Given'].reset_index()
negative_rating_of_reviewers.rename(columns={'Total_Number_of_Reviews_Reviewer_Has_Given':'count_Total_Number_of_Reviews_Reviewer_Has_Given'},inplace = True)
negative_rating_of_reviewers = negative_rating_of_reviewers[(negative_rating_of_reviewers['count_Total_Number_of_Reviews_Reviewer_Has_Given'] < 50) & (negative_rating_of_reviewers['Reviewer_Score'] > 6.0) & (negative_rating_of_reviewers['Average_Score'] < 6.0)].sort_values('Reviewer_Score', ascending =False)
negative_rating_of_reviewers
```

```python
positive_rating_of_reviewers.merge(hotels, on='Hotel_Name').drop_duplicates(['Average_Score','Reviewer_Score', 'Reviewer_Nationality'])[['Hotel_ID', 'Hotel_Name', 'Average_Score', 'Reviewer_Score','Reviewer_Nationality' ]]
```

```python
prw = positive_rating_of_reviewers.pivot_table(index ='Hotel_Name', columns = 'Reviewer_Nationality', values = 'Average_Score')
prw
```

```python
prw.fillna(0, inplace = True)
prw
```
